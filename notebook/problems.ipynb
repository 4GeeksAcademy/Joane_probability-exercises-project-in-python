## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO

import numpy as np
import random

def roll_dice(n_simulations=1000):
    count = 0

    for i in range(n_simulations):
        # Roll each dice
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
    
        # Total of the value to get score
        score = die1 + die2

        # If we were to add the score?
        if (score % 2 == 0) or (score > 7):
            count += 1

    # Calculate probability
    return count / n_simulations

# Now call the function and print the result
probability = roll_dice() * 100  # This will give us the percentage
print(f"The probability of rolling an even number or greater than 7 is {np.round(probability, 2)}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np

# Define the number of balls of each color
n_white = 10
n_red = 20
n_green = 30

# Define the total number of balls
n_total = n_white + n_red + n_green

# Define the desired number of draws and the desired outcomes
n_draws = 5
outcome1 = (3, 2)  # 3 white, 2 red
outcome2 = (5, 0, 0)  # All the same color (any color)

# Run the experiment 1000 times
n_experiments = 1000
success_count1 = 0
success_count2 = 0

for _ in range(n_experiments):
  # Draw 5 balls with replacement
  drawn_balls = np.random.choice([0, 1, 2], size=n_draws, p=[n_white/n_total, n_red/n_total, n_green/n_total], replace=True)

  # Count successes for outcome 1
  if np.sum(drawn_balls == 0) == outcome1[0] and np.sum(drawn_balls == 1) == outcome1[1]:
    success_count1 += 1

  # Count successes for outcome 2
  if len(np.unique(drawn_balls)) == 1:
    success_count2 += 1

# Calculate probabilities
probability1 = success_count1 / n_experiments
probability2 = success_count2 / n_experiments

# Print the results
print("Probability of getting 3 white and 2 red balls:", probability1)
print("Probability of getting all balls of the same color:", probability2)

```
